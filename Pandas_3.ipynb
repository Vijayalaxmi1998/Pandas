# Acessing Data from DataFrame

```python
import pandas as pd
olympic_data=pd.read_csv("https://raw.githubusercontent.com/svkarthik86/Advanced-python/main/olympics.csv",skiprows=4)
olympic_data.head(1)
```

## Accessing column using []

```python
olympic_data["Edition"]
```

```python
type(olympic_data["Edition"])
```

```python
olympic_data[["Edition","Sport","NOC"]]
```

```python
type(olympic_data[["Edition","Sport","NOC"]])
```

## Using.object

```python
olympic_data.Edition
```

```python
## we can access only one columns in data series
```

## using : loc
- String based index

```python
olympic_data.loc[0:100,"Edition"]
```

```python
olympic_data.loc[0:100,["NOC","Edition","Sport"]]
```

```python
olympic_data.loc[0:100,"City":"Event"]
```

```python
olympic_data.loc[0:100:2,"City":"Event":2]    # it include the last column and index
```

## using iloc
- integer based, numerical index

```python
olympic_data.iloc[0:100,0:3]    # it not included last column and 
```

```python
olympic_data.iloc[0:100,2:3]
```

```python
type(olympic_data.iloc[0:100,2:3])
```
